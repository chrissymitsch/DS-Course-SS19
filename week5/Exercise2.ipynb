```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import glob
```

```python
data = []
list = [pd.read_json(filename, lines=True) for filename in glob.glob("../week4/data/return-data/*.txt")]
df = pd.concat(list, axis=0)
df.head()
```

```python
df.info()
```

```python
zipCode_dummies = pd.get_dummies(df['zipCode'])
df.drop(['basket','transactionId','zipCode'],axis=1,inplace=True)
df = pd.concat([df,zipCode_dummies],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['returnLabel'],axis=1)
y = df['returnLabel']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)

predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```
